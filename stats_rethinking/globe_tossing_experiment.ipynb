## Model settings


$$
\begin{array}{l}
p\ \varpropto Uniform( 0,\ 1)\\
W\ \varpropto Binom( N,\ p)\\
N = W + L\\
\\
\\
Joint\ Probability\ of\ all\ the\ parameters\ involved\ are \\
Pr(W,L,p) = Pr(W,L|p) * Pr(p) \\
\end{array}
$$

$Pr(W,L|p)$ is the joint probability of Water and Land conditioned on any value of $p$

This can be also written as the other way of $p$ is conditioned on Given value of W and L.

$ Pr(W,L,p) = Pr(p|W,L) * Pr(W,L)$

Every settings like this, the joint probability of all the variables turns into bayes theorm, from that we will do the inference on the unobserved variable, here it's $p$.

$$
\\
Pr(p|W,L) * Pr(W,L) = Pr(W,L|p) * Pr(p)
\\
Pr(p|W,L) = \frac{Pr(W,L|p) * Pr(p)}{Pr(W,L)}
\\
Posterior = \frac{Probability\ of\ the\ Data \times\ Prior}{Average\ Probability\ of\ Data}
$$

Here the $P(W,L)$ is the evidence, or mariginalised likelihood. We can write this in using likelihood term by margenlize it.

$P(W,L) = \int Pr(W,L|p)  Pr(p)  dp = E(Pr(W,L|p))$
