```python
import pandas as pd
import numpy as np
```

```python
expense_df = pd.read_csv(
    "/Users/haridas/Downloads/walnut_expense_report_01-05-2015_to_30-06-2020.csv"
)
```

```python
expense_df["AMOUNT"] = expense_df.AMOUNT.apply(lambda x: float(str(x).replace(",", "")))
```

```python
expense_df["YEAR"] = expense_df.DATE.apply(lambda x: str(x).split("-")[-1])
```

```python
expense_only_df = expense_df[expense_df.EXPENSE == "Yes"]
```

```python
expg = expense_only_df.groupby("YEAR").sum()["AMOUNT"]
```
